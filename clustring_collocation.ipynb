# En notebook for å lage clustre fra kollokasjoner

Start med å importere de kommandoene som trengs fra nbtext

```python
import nbtext as nb
from nbtext import cloud, get_urn, Cluster, Corpus
%matplotlib inline
```

### Definer et cluster for et ord over en gitt periode fra bøker eller aviser

```python
acluster = Cluster(
    'demokrati', 
    period=(1970, 1980), 
    before=5, 
    after=5, 
    corpus='avis',
    reference=150, 
    word_samples=500)
```

```python
acluster50 = Cluster(
    'demokrati', 
    period=(1950, 1960), 
    before=5, 
    after=5, 
    corpus='avis',
    reference=150, 
    word_samples=500)
```

```python
bcluster = Cluster(
    'sannhet', 
    period=(1970, 1980), 
    before=5, 
    after=5, 
    corpus='bok',
    reference=150, 
    word_samples=500)
```

```python
acluster.collocates.sort_values(by='demokrati', ascending=False)[0:5]
```

```python
acluster.reference.sort_values(by='reference_corpus', ascending=False)[0:5]
```

```python
acluster50.search_words('demokrati diktatur . ,')
```

```python
bcluster.cluster_set()
```

```python
acluster.cluster_set()
```

Clustret kan studeres som en ordsky

```python
acluster.cluster_set(aslist=False)
```

```python
nb.cloud(acluster.cluster_set(aslist=False)[:150])
```

```python
nb.cloud(acluster50.cluster_set(aslist=False)[:150])
```
