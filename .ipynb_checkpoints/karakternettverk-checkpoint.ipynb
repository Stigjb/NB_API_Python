## Finn bøker

Benytt kommandoen _get_urn()_ for å hente urner basert på kategorier av bøker

```python
# Hent URN-er, lag korpus med get_urn()

urner = get_urn({
    'author':'masterson%lou%', 
    'year':1900, 
    'neste':160,
    'limit':100,
    'title':'%kane%'
    
})

print('Antall verk: ', len(urner))
```

```python
def navn(urn):
    import requests
    if type(urn) is list:
        urn = urn[0]
    r = requests.get('https://api.nb.no/ngram/tingnavn', json={'urn':urn})
    return r.json()
```

```python
kane = navn(urner[10])
```

```python
kanetext = nb.get_freq(urner[0][0])
```

```python
kanetext
```

```python
kaneord = [n[0] for n in kane if n[1] > 1]
```

```python
# Vis flere treff med urner[fra:til]
for i in range(len(urner)):
    print(i, urner[i])
```
